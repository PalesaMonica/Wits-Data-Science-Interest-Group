Understanding Data Types in Your Dataset

When working with data , it’s important to know the types of data you have because it affects how you analyze and manipulate it. Most datasets include three main types:
1. Categorical Data
   -  Represents discrete labels or categories.
   - Examples: Gender (Male/Female), Country (South Africa, USA), Product Type (A, B, C).
   - Often stored as strings (object) in Pandas, but can also be converted to category type for efficiency.


2. Numeric Data
  -  Represents numbers, either integers or floats.
  -  Examples: Age (25, 30), Salary (50000.0, 75000.5), Temperature (23.5).
  -  Used for calculations, statistics, and plotting.

3. Datetime Data
    - Represents dates and times.
    - Examples: 2026-02-26, 2026-02-26 14:30:00.
    - Important for time series analysis, sorting, and filtering by dates.

## SQL Basics

SQL (Structured Query Language) is the standard language used to interact with relational databases like PostgreSQL. It allows you to retrieve, insert, update, and delete data in a structured way.

Selecting Data

Retrieve data from a table using SELECT

```SQL
-- Select all columns
SELECT * FROM employees;

-- Select specific columns
SELECT name, age, salary FROM employees;

```
Filtering Data

Use WHERE to filter rows based on conditions.
```SQL
SELECT * FROM employees
WHERE age > 30;

SELECT * FROM employees
WHERE department = 'Sales';

```
Sorting Data

Use ORDER BY to sort results.

```SQL
SELECT * FROM employees
ORDER BY salary DESC;  -- Descending order
```

Aggregating Data

Use functions like COUNT, SUM, AVG, MIN, MAX.

```SQL
SELECT COUNT(*) FROM employees;       -- Total employees
SELECT AVG(salary) FROM employees;    -- Average salary
SELECT MIN(age), MAX(age) FROM employees; -- Youngest & oldest
```




### How to use pandas with PostgreSQL
Pandas can directly interact with PostgreSQL databases, allowing you to query data and load it into a DataFrame for analysis. This is very useful when working with real datasets stored in a database.

````python
from sqlalchemy import create_engine
````

````python
engine = create_engine(
    "postgresql+psycopg2://username:password@127.0.0.1:5432/database"
)
````

````python
try:
    conn = engine.connect()
    print("SQLAlchemy connection successful!")
    conn.close()
except Exception as e:
    print("Connection failed:", e)
````

````python
df = pd.read_sql("SELECT * FROM film;", engine)
````
